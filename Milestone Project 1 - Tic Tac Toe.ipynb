```python
def game_instructions():
    board = range(1,10)
    print("Welcome to Tic Tac Toe!!\n")
    print("The board is based on the position of the number keypad on the keyboard.\n")
    print(f'{board[6]} | {board[7]} | {board[8]}')
    print("--|---|--")
    print(f'{board[3]} | {board[4]} | {board[5]}')
    print("--|---|--")
    print(f'{board[0]} | {board[1]} | {board[2]}\n')
    print("Each player should use the numbers to place an 'X' or 'O'.\n")
    print("-------------------------------------------------------------------------------")
    
```

```python
game_instructions()
```

```python
def player1_choice():
    # A function that helps validate player 1's choice. 
    
    while True: 
        player1 = input("Player 1, Do you want to be 'X' or 'O': ")
        if player1 not in ["X","O"]:
            print("Try again. Please choose if you want to be 'X' or 'O'\n")
        else:
            print(f'Player 1 is {player1}')
            break 
            
    return player1
          
```

```python
def player2_choice(player1):
    # A function that automatically selects player 2's choice based on player 1's choice
    
    if player1 == "X":
        player2 = "O"
        print(f'Player 2 is {player2}\n')
        
    else:
        player2 = "X"
        print(f'Player 2 is {player2}\n')
        
    return player2
    
    
```

```python
board = [" "," " ," "," "," "," "," "," "," "," "]
def display_board():
    print("\n") 
    print(f'{board[6]} | {board[7]} | {board[8]}')
    print("--|---|--")
    print(f'{board[3]} | {board[4]} | {board[5]}')
    print("--|---|--")
    print(f'{board[0]} | {board[1]} | {board[2]}')
    print("\n") 
    
    
```

```python
def line_check(char,box1,box2,box3):
    if char == board[box1] and char == board[box2] and char == board[box3]:
        return True 
```

```python
def winner_check(char): 
    # Row Checks
    if line_check(char,6,7,8):
        return True
    if line_check(char,3,4,5):
        return True
    if line_check(char,0,1,2):
        return True
    
    # Column Check 
    if line_check(char,0,3,6):
        return True
    if line_check(char,7,4,1):
        return True
    if line_check(char,2,5,8):
        return True
    
    # Diagonal Check
    if line_check(char,0,4,8):
        return True
    if line_check(char,6,4,2):
        return True
```

```python
# Tic Tac Toe Game Logic 

# 1.) Start the game
# Game instructions 
# Ask if player one wants to be 'X' or 'O' 
# Based on player 1 input, player 2 will be automatically selected

restart = "Y"
while restart.lower() == "y":
    game_instructions()
    player1 = player1_choice()
    player2 = player2_choice(player1)

    # 2.) Display the Board
    board = [" "," "," "," "," "," "," "," "," "," "]
    display_board()
    start_game = True 
    turn_counter = 0 


    while start_game:
    
        while True:
        
        # 3.) Collect Player 1 input 
        # Collect the positional value entered by player 1
        # Check if other values are in the same position 
        # Place player's item in position
        # Display the board
    
            player1_position = input(f'{player1}, Choose a position on the board -: ')
            if player1_position.isdigit():
                player1_position = int(player1_position)
                if player1_position >= 1 and player1_position <= 9:
                    if board[player1_position - 1] != "X" and board[player1_position-1] != "O":
                        board[player1_position - 1] = player1
                        turn_counter += 1
                        display_board()
                        break 
                
                    else:
                        print("This position is taken\n")
                        display_board()
                else:
                    print("Please enter a valid position\n")
                    display_board()
            
            else:
                print("Please enter a valid position\n")
                display_board()
          
        # Check if Player 1 has won 
        if winner_check(player1) == True:
            print(f'{player1} has won\n')
            break
     
        # After 9 turns, if nobody has one -: The game will end as a draw
        if turn_counter == 9:
            print("The game is a draw\n")
            break
        
        while True:
        
        # 4.) Collect Player 2 input 
        # Collect the positional value entered by player 2
        # Place player's item in position 
        # Display the board
    
            player2_position = input(f'{player2}, Choose a position on the board -: ')
            if player2_position.isdigit():
                player2_position = int(player2_position)
                if player2_position >= 1 and player2_position <= 9:
                    if board[player2_position - 1] != "X" and board[player2_position - 1] != "O":
                        board[player2_position - 1] = player2
                        turn_counter += 1
                        display_board()
                        break
                
                    else:
                        print("This position is taken\n")
                        display_board()
                else:
                    print("Please enter a valid position\n")
                    display_board()
            
            else:
                print("Please enter a valid position\n")
                display_board()
            
        # Check if Player 2 has won
    
        if winner_check(player2) == True:
            print(f'{player2} has won\n')
            break
        
    
    # 6.) Ask players if they want to play again? 
    
    manual_restart = input("Would you like to play again. Press 'Y' to continue or 'N' to finish -: ")
    print("\n")
    if manual_restart.lower() not in ["y","n"]:
        while True: 
            print("Please Press 'Y' to continue or 'N' to finish -: \n")
            manual_restart = input("Would you like to play again. Press 'Y' to continue or 'N' to finish -: ")
            print("\n")
            
            if manual_restart.lower() == "n":
                print("Thank you for playing!!!!")
                restart = manual_restart
                break
            
            elif manual_restart.lower() == "y":
                break
            
            else:
                pass
    

    elif manual_restart.lower() == "n":
            print("Thank you for playing!!!!")
            restart = manual_restart
            

```
